# VEClim Tutorials Hub

Welcome to the **VEClim Tutorials Hub**, the main entry point to modelling tutorials,
training materials, and educational resources co-developed by the VEClim team and
the climate-sensitive vector dynamics modelling community. These resources support students, researchers, and public‑health
professionals working on **vector activity prediction** and **early‑warning system design**.

---

## 🌍 About VEClim
VEClim is an evolving platform that integrates climate data, mechanistic modelling,
and prediction tools to help characterise and anticipate vector‑borne disease risk.

We begin VEClim’s tutorial library with the full set of materials from the
**Climate‑Sensitive Vector Dynamics Modelling Workshop – Nicosia 2025**, with
additional tutorials to be added as the platform grows.

---
