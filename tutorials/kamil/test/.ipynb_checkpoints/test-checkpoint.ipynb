```r
source("population.R")
```

```r
prepare <- function(filename) {
    return(system(sprintf("node compile.js %s", filename), intern = TRUE))
}
```

```r
prepare("ex_001")
```

```r
model <- PopulationModel$new("ex_001.dylib")
```

```r
ret <- model$sim(ftime = 20, 
                 y0 = list(larva = 100.0))

plot(ret$ret[1, , model$popids[["larva"]]], t = "l")
```

```r
model <- PopulationModel$new("ex_003.dylib")
```

```r
temp <- seq(0.0, 40.0, length.out = 1000)

param <- model$param

ret <- model$sim(
  ftime = length(temp),
  envir = list(
    temp = temp
  ),
  pr = param,
  rep = -1
)

plot(head(temp,-1), 
     ret$iret[1, , model$intids[["int_larva_dev_mean"]]], 
     t = "l", 
     ylim = c(0, 100))
```

```r
temp <- rep(13.0, 200)

param <- model$param

ret <- model$sim(
  ftime = length(temp),
  envir = list(
    temp = temp
  ),
  y0 = list(
    larva = 100.0
  ),
  pr = param,
  rep = 1
)

plot(ret$ret[1, , model$popids[["larva"]]], t = "l")
lines(cumsum(ret$iret[1, , model$intids[["pupa"]]]), t = "l")
```
